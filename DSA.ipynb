```python
def binary_search(a,s):
  if(len(a)==0):
    return("element not found")
  if(len(a)==1):
    if(a[0]==s):
      return("element found")
    else:
      return("element not found")

  if(a[int(len(a)/2)]==s):
    return("element found")
  elif(a[int(len(a)/2)]>s):
    return(binary_search(a[:int(len(a)/2)],s))
  else:
    return(binary_search(a[int(len(a)/2)+1:],s))

```

```python
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    mid = 0

    while low <= high:

        mid = (high + low) // 2

        # If x is greater, ignore left half
        if arr[mid] < x:
            low = mid + 1

        # If x is smaller, ignore right half
        elif arr[mid] > x:
            high = mid - 1

        # means x is present at mid
        else:
            return mid

    # If we reach here, then the element was not present
    return -1


# Test array
```

# merge sort with recursion
-divide and concer

```python
def merge_sort(a):
  if(len(a)==1):
    return(a)
  else:
    m=int(len(a)/2)
    r1=merge_sort(a[:m])
    r2=merge_sort(a[m:])
    p1=0
    p2=0
    out=[]
    while(p1!=len(r1) and p2!=len(r2)):
      if(r1[p1]>r2[p2]):
        out.append(r2[p2])
        p2=p2+1
      else:
        out.append(r1[p1])
        p1=p1+1
    if(p1!=len(r1)):
      out.extend(r1[p1:])
    if(p2!=len(r2)):
      out.extend(r2[p2:])
    return(out)

```

```python
import copy
```

```python
def pivote(l):
  a=copy.copy(l)
  s=1
  p=0
  e=len(a)-1
  while(s<e):

    while(a[p]>a[s] and s!=e):
      s=s+1
    while(a[p]<a[e] and e!=s):
      e=e-1
    if(s<e):
      a[s],a[e]=a[e],a[s]
  # print(f"pointer-{p},start-{s},end-{e},array-{a}")
  if(a[s]>=a[p]):
    a[s-1],a[p]=a[p],a[s-1]
    p=s-1
  elif(a[e]<a[p]):
    a[e],a[p]=a[p],a[e]
    p=e

  return(a,p)
```

```python
def quick_sort(l):
  a=copy.copy(l)
  if(len(a)<2):
    return(a)
  else:
    arr,i=pivote(a)
    x=quick_sort(arr[0:i])
    y=quick_sort(arr[i+1:])
    x.append(arr[i])
    x.extend(y)
    return(x)
```

```python
def subset(a):
  if len(a)==1:
    return([a])
  else:
    # x=[]
    x=(subset(a[:-1]))
    y=a[-1]
    temp=x.copy()
    for i in temp:
      t1=i.copy()
      # print(t1)
      t1.append(y)
      x.append(t1)
    x.append([y])
    return(x)

```

## maize problem

```python
def maize(path,cp):
  if(cp==22):
    return([[]],22)
  else:
    out=[]
    if((cp%10)<2):
      left=maize(path,cp+1)[0]
      for i in left:
        i.append(cp+1)
        out.append(i)
    if (int(cp/10)<2):
      right=maize(path,cp+10)[0]
      for i in right:
        i.append(cp+10)
        out.append(i)
    if((cp%10)<2 and int(cp/10)<2):
      cross=maize(path,cp+11)[0]
      for i in cross:
        i.append(cp+11)
        out.append(i)
    return(out,22)
```
